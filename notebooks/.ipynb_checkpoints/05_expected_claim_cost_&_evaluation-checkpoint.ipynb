## Chapter 5 Overview

This chapter integrates the frequency and severity models to estimate expected claim cost under the actuarial pricing framework. We will:

- Compare frequency models (Poisson vs. Negative Binomial) and select the most appropriate approach for handling dispersion.  
- Evaluate severity models (Gamma, Lognormal, GPD, and Hybrid) to assess fit across ordinary and extreme claims.  
- Combine frequency and severity outputs to approximate expected claim cost using representative datasets.  
- Visualize the expected loss distribution and examine residual diagnostics to validate model performance.  
- Analyze feature importance from Random Forest, XGBoost, and GLM to identify key business drivers.  
- Conclude with a business summary highlighting implications for pricing, reserving, and risk management.  

**Note:** Because the frequency and severity datasets are independent, expected loss will be computed at the portfolio level rather than policy‑by‑policy. This preserves the actuarial framework but does not capture dependence between claim occurrence and claim size.


```python
# Data handling
import pandas as pd
import numpy as np

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns

# System
import os

```

```python
# Compare Frequency Models in a Table
# Load metrics
poisson_metrics = pd.read_csv("../data/poisson_metrics.csv")
nb_metrics = pd.read_csv("../data/nb_metrics.csv")

# Combine into comparison table
freq_comparison = pd.concat([poisson_metrics.assign(Model="Poisson"),
                             nb_metrics.assign(Model="Negative Binomial")],
                            ignore_index=True)

   
freq_comparison
```

### Interpretation – Frequency Models
Both Poisson and Negative Binomial models produced nearly identical error metrics (MSE, MAE, RMSE). This suggests that overdispersion in the data is modest, so the added flexibility of the Negative Binomial does not yield a large improvement in predictive accuracy. However, the NB model remains preferable in practice because it can handle potential overdispersion more robustly, making it a safer choice for claim frequency modelling.


```python
gamma_metrics = pd.read_csv("../data/gamma_metrics.csv")
lognormal_metrics = pd.read_csv("../data/lognormal_metrics.csv")
gpd_metrics = pd.read_csv("../data/gpd_metrics.csv")
hybrid_metrics = pd.read_csv("../data/hybrid_metrics.csv")

sev_comparison = pd.concat([
    gamma_metrics.assign(Model="Gamma"),
    lognormal_metrics.assign(Model="Lognormal"),
    gpd_metrics.assign(Model="GPD"),
    hybrid_metrics.assign(Model="Hybrid (Lognormal+GPD)")
], ignore_index=True)

sev_comparison.set_index("Model", inplace=True)
sev_comparison
```

### Interpretation – Severity Models
- **Gamma**: Achieved relatively lower error metrics compared to Lognormal, but its AIC was much higher, indicating weaker overall fit.  
- **Lognormal**: Showed higher error metrics than Gamma, but a substantially lower AIC, suggesting better likelihood-based fit for the bulk of claims.  
- **GPD**: Captured extreme tail behavior but had very high error metrics, reflecting poor fit for ordinary claims.  
- **Hybrid (Lognormal+GPD)**: Balanced performance by combining Lognormal for bulk claims and GPD for tails. Error metrics were close to Lognormal, while residual plots and Q‑Q plots confirmed improved tail fit.  

Overall, the Hybrid model provides the most practical severity fit, as it captures both central tendency and extreme losses more effectively than single-distribution models.



```python
# Residuals for hybrid model
df_combined = pd.read_csv("../data/severity_predictions.csv")
df_combined['residuals'] = df_combined['total_claim_amount'] - df_combined['predicted']

plt.figure(figsize=(8,6))
sns.scatterplot(x=df_combined['predicted'], y=df_combined['residuals'], alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals - Hybrid Severity Model")
plt.xlabel("Predicted Severity")
plt.ylabel("Residuals")
plt.show()
```

### Interpretation - residuals for hybrid model

The residuals are generally centered around zero, which suggests the model is not systematically over‑ or under‑predicting across the range of predicted severities.

Variance appears relatively stable across most of the prediction range, though there is some widening at higher predicted severities — a common sign of heavier tails in claim data.

This pattern supports the use of the hybrid approach: the Lognormal handles the bulk of claims well, while the GPD improves fit in the tail, reducing bias for extreme losses.

```python
# Compute Expected Claim Cost

# Load predictions
freq_preds = pd.read_csv("../data/nb_frequency_predictions.csv")
sev_preds = pd.read_csv("../data/severity_predictions.csv")

# Compute averages
avg_freq = freq_preds['NB_Predictions'].mean()
avg_sev = sev_preds['predicted'].mean()

expected_claim_cost = avg_freq * avg_sev
print(f"Expected Claim Cost: {expected_claim_cost:.2f}")
```

### Interpretation – Expected Claim Cost
The expected claim cost of **2,760.99** represents the average portfolio risk when combining the Negative Binomial frequency model with the Hybrid severity model. This value reflects the typical financial exposure per policy, balancing both ordinary claims and extreme tail events. It provides a practical benchmark for pricing and reserving decisions, ensuring that premiums and capital allocation align with the underlying risk profile.


```python
# Visualize Expected Loss Distribution - Histogram of predicted severities × frequencies.

# Simulate expected losses at portfolio level
expected_losses = freq_preds['NB_Predictions'] * sev_preds['predicted']

plt.figure(figsize=(10,6))
sns.histplot(expected_losses, bins=50, kde=True, color="skyblue")
plt.title("Expected Loss Distribution")
plt.xlabel("Expected Claim Cost")
plt.ylabel("Frequency")
plt.show()
```

### Interpretation – Expected Loss Distribution
The histogram shows a right‑skewed distribution of expected claim costs. Most predicted losses cluster at lower values, reflecting the predominance of ordinary claims. However, the long tail indicates that a small number of extreme claims contribute disproportionately to overall risk. This pattern highlights the importance of tail modelling (via GPD in the hybrid severity model) and confirms that portfolio risk is driven not only by frequent moderate claims but also by rare, high‑severity events.


```python
# Helper function for plotting
def plot_feature_importance(df, feature_col, value_col, title, color):
    plt.figure(figsize=(10,6))
    plt.barh(df[feature_col], df[value_col], color=color)
    plt.xlabel(value_col)
    plt.title(title)
    plt.gca().invert_yaxis()  # highest at top
    
    # Add value labels
    for i, v in enumerate(df[value_col]):
        plt.text(v + 0.001, i, f"{v:.3f}", va='center')
    
    plt.show()

# --- Random Forest ---
rf_importance_df = pd.read_csv("../data/rf_feature_importance.csv")
rf_top10 = rf_importance_df.sort_values(by="Importance", ascending=False).head(10)

plot_feature_importance(rf_top10, "Feature", "Importance",
                        "Random Forest - Top 10 Features", "skyblue")

# --- XGBoost ---
xgb_importance_df = pd.read_csv("../data/xgb_feature_importance.csv")
xgb_top10 = xgb_importance_df.sort_values(by="Importance", ascending=False).head(10)

plot_feature_importance(xgb_top10, "Feature", "Importance",
                        "XGBoost - Top 10 Features", "orange")

# --- GLM Coefficients ---
glm_coef_df = pd.read_csv("../data/glm_coefficients.csv")

# Positive predictors
glm_top_positive = glm_coef_df.sort_values(by="Coefficient", ascending=False).head(10)
plot_feature_importance(glm_top_positive, "Feature", "Coefficient",
                        "GLM - Top Positive Predictors", "green")

# Negative predictors
glm_top_negative = glm_coef_df.sort_values(by="Coefficient", ascending=True).head(10)
plot_feature_importance(glm_top_negative, "Feature", "Coefficient",
                        "GLM - Top Negative Predictors", "red")

```

### Interpretation – Feature Importance
The three models highlight consistent drivers of claim severity:

Random Forest: Subscription length, vehicle age, and customer age dominate, showing that policy tenure and demographic/vehicle characteristics strongly influence risk.

XGBoost: Vehicle age, mechanical attributes (torque, displacement, steering type), and customer age emerge as key predictors, adding more detail on how vehicle specifications affect claims.

GLM: Subscription length and certain regional codes are the strongest positive predictors, reinforcing the importance of tenure and geographic risk factors.

Across all models, subscription length, vehicle age, and customer age repeatedly appear as top features, confirming their central role in explaining claim variability. The consistency across methods strengthens confidence in these variables as primary business drivers for pricing and reserving decisions.

## Business Summary

- **Frequency Models**: Both Poisson and Negative Binomial produced similar error metrics, but the Negative Binomial is preferred for its ability to handle overdispersion more robustly. This ensures claim frequency estimates remain stable even when variance exceeds the mean.

- **Severity Models**: Gamma and Lognormal each captured different aspects of claim severity, while GPD alone fit the tail but poorly represented ordinary claims. The Hybrid (Lognormal + GPD) model provided the best balance, capturing both central tendency and extreme losses.

- **Expected Claim Cost**: The combined NB frequency and Hybrid severity model yielded an expected claim cost of **2,760.99**, representing the average portfolio risk per policy. This value is a practical benchmark for pricing and reserving.

- **Loss Distribution**: The expected loss distribution is right‑skewed, confirming that most claims are moderate but a small number of extreme claims drive tail risk. This validates the importance of tail modelling for capital allocation.

- **Residual Diagnostics**: Residuals were centered around zero with relatively stable variance, showing no major bias. Q‑Q plots confirmed improved tail fit under the Hybrid model, strengthening confidence in its robustness.

- **Feature Importance**: Subscription length, vehicle age, and customer age consistently emerged as top predictors across Random Forest, XGBoost, and GLM. Regional codes and vehicle specifications also contributed, but with smaller influence. These findings highlight tenure, demographics, and vehicle characteristics as key business drivers.

### Business Impact
This modelling framework supports:
- **Fairer pricing** by aligning premiums with true risk drivers.  
- **Better capital allocation** by accounting for both frequent moderate claims and rare extreme losses.  
- **Improved risk management** through robust frequency and severity modelling, ensuring portfolios are resilient to tail events.  

Overall, the NB + Hybrid model combination provides a transparent, data‑driven foundation for decision‑making in insurance pricing and reserving.
